```python
def print_uni_col_values(df1):
    for column in df1:
        if df1[column].dtypes == 'object':
            print(f'{column} : {df1[column].unique()}')
```

```python
print_uni_col_values(df1)
```

```python
df1.replace('No phone service' , 'No' , inplace = True)
df1.replace('No internet service' , 'No' , inplace = True)
```

```python
print_uni_col_values(df1)
```

```python
yes_no_columns = ['Partner' , 'Dependents' , 'PhoneService' , 'MultipleLines' , 'OnlineSecurity' , 'OnlineBackup','DeviceProtection' , 'TechSupport' , 'StreamingTV' , 'StreamingMovies' , 'PaperlessBilling' , 'Churn']

for col in yes_no_columns:
    df1[col].replace({'Yes':1 , 'No':0} , inplace = True)
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1 , 'Male':0} , inplace = True)
```

```python
df1.gender.unique()
```

```python
df2 = pd.get_dummies(data=df1 , columns = ['InternetService' , 'Contract' , 'PaymentMethod']).astype(int)
df2.columns
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
df2.head(
)
```

```python
cols_to_scale = ['tenure' , 'MonthlyCharges' , 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(5)
```

```python
X = df2.drop('Churn' ,axis=1)
Y= df2.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2 , random_state=2)
```

```python
X_train.shape
```

```python
X_train[:5]
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20 , input_shape=(26,) , activation = 'relu'),
    keras.layers.Dense(15 , activation = 'relu'),
    keras.layers.Dense(10 , activation = 'relu'),
    keras.layers.Dense(5 , activation = 'relu'),
    keras.layers.Dense(1 , activation = 'sigmoid')
])

model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = 'accuracy')
model.fit(X_train , Y_train , epochs =100)
```

```python
model.evaluate(X_test, Y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:5]
```

```python
Y_test[:5]
```

```python
yp =[]
for element in y_pred:
    if element>0.5:
        yp.append(1)
    else:
        yp.append(0)
```

```python
yp[:5]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
import seaborn as sns
cm = tf.math.confusion_matrix(labels = Y_test , predictions = yp)
sns.heatmap(cm , annot=True , fmt ='.1f')
plt.xlabel('Preidcted')
plt.ylabel('Truth')
```
